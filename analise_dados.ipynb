# INTRODUÇÃO

Este notebook tem como objetivo realizar uma análise exploratória de dados (EDA) em um conjunto de dados de pesquisa sobre profissionais da área de dados. Através desta análise, buscaremos entender melhor as características demográficas, profissionais e salariais desses indivíduos. Serão utilizadas as bibliotecas Pandas e NumPy para manipulação e cálculo de estatísticas básicas, e Matplotlib para visualização.

```python
# Definindo uma lista de idades
# Este código cria uma lista Python simples chamada `lista_idades` contendo alguns valores numéricos que representam idades.
lista_idades = [26,30,32,22,26,35,40,20,43,31,23]
```

```python
# Calculando a mediana da lista de idades
# Utiliza a função `np.median()` do NumPy para calcular a mediana da `lista_idades`.
# A mediana é o valor do meio em um conjunto de dados ordenado, sendo menos sensível a _outliers_ do que a média.
mediana=np.median(lista_idades)
print("Mediana:", mediana)
```

```python
# Calculando o Intervalo Interquartil (IQR)
# IQR é a diferença entre Q3 e Q1 e representa a dispersão dos 50% centrais dos dados.
IQR = Q3 - Q1
IQR
```

```python
# Calculando o limite superior para identificação de outliers usando o método IQR
# Outliers são valores acima de Q3 + (1.5 * IQR)
lim_superior = Q3 + (1.5*IQR)
lim_superior
```

```python
# Calculando o limite inferior para identificação de outliers usando o método IQR
# Outliers são valores abaixo de Q1 - (1.5 * IQR)
lim_inferior = Q1 - (1.5*IQR)
lim_inferior
```

#Correlação

```python
# Calculando a correlação de Pearson entre as colunas 'IDADE' e 'SALARIO'
# Este coeficiente mede a força e a direção de uma relação linear entre duas variáveis contínuas.
correlacao_continua = dados['IDADE'].corr(dados['SALARIO'])
correlacao_continua
```

```python
# Importando a função 'chi2_contingency' do módulo 'scipy.stats'
# Esta função é usada para realizar o teste qui-quadrado de independência.
from scipy.stats import chi2_contingency
```

```python
# Definindo uma função para calcular o coeficiente V de Cramer
# O V de Cramer é uma medida da força de associação entre duas variáveis nominais.
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1,coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape)-1
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
```

```python
# Calculando o coeficiente V de Cramer entre 'COR/RACA/ETNIA' e 'NIVEL DE ENSINO'
# Isso mede a associação entre essas duas variáveis categóricas.
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
```

```python
# Calculando o coeficiente V de Cramer entre 'GENERO' e 'NIVEL DE ENSINO'
# Isso mede a associação entre o gênero e o nível de ensino dos participantes.
cramer_coeficiente(dados['GENERO'], dados['NIVEL DE ENSINO'])
```

```python
# Criando uma tabela de contingência (crosstab) entre 'COR/RACA/ETNIA' e 'NIVEL DE ENSINO'
# Esta tabela mostra a frequência conjunta de cada combinação das categorias das duas variáveis.
tabela_cruzada = pd.crosstab(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
tabela_cruzada
```
